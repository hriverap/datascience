* The aim is analyze a file in audio format and build a cloud words.

** Steps: **
* Load libraries
* Load audio file
* Extract text from audio file

```python
import speech_recognition as sr
import numpy as np
import matplotlib.pyplot as plt
import requests

from PIL import Image
from wordcloud import WordCloud
```

* Activamos la API de reconocimiento

```python
r = sr.Recognizer()
```

```python
harvard = sr.AudioFile("./data/harvard.wav")
```

```python
with harvard as source:
    audio = r.record(source)
```

```python
type(audio)
```

* Extraemos el texto desde el audio

```python
text = r.recognize_google(audio)
```

```python
text
```

* Generamos el cloud de palabras para el texto

```python
mask = np.array(Image.open("./data/twitter_mask.png"))
```

```python
word_cloud = WordCloud(width = 300, height = 300, background_color='white', mask=mask).generate(text)
```

```python
plt.figure(figsize=(10,8), facecolor = 'white', edgecolor='blue')
plt.imshow(word_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
